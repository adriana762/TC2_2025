Comienzo hallando la funcion de transferencia del circuito:
$$
(V_x - V_o) G_2 = V_o C_2
$$

Despejamos $ V_x $:

$$
V_x G_2 - V_o G_2 = V_o C_2 \implies V_x G_2 = V_o (C_2 + G_2) \implies
$$

$$
V_x = \frac{C_2 + G_2}{G_2} V_o
$$

La segunda ecuación es:

$$
(V_i - V_x) G_1 = (V_x - V_o) G_2 + (V_x - V_o) C_1
$$

$$
V_i G_1 - V_x G_1 = (V_x - V_o)(G_2 + C_1)
$$

$$
V_i G_1 = V_x G_1 + (V_x - V_o)(G_2 + C_1)
$$

$$
V_i G_1 = V_x G_1 + V_x (G_2 + C_1) - V_o (G_2 + C_1)
$$

$$
V_i G_1 + V_o (G_2 + C_1) = V_x (G_1 + G_2 + C_1)
$$

Ahora reemplazamos $V_x $ con la expresión hallada:

$$
V_i G_1 + V_o (G_2 + C_1) = \frac{C_2 + G_2}{G_2} V_o (G_1 + G_2 + C_1)
$$

$$
V_i G_1 = V_o \left[ \frac{C_2 + G_2}{G_2} (G_1 + G_2 + C_1) - (G_2 + C_1) \right]
$$

Por lo tanto, la función transferencia es:

$$
\boxed{H(s) = \frac{V_o}{V_i} = \frac{\frac{G_1 G_2}{C_1 C_2}}{s^2 + s \frac{G_1 + G_2}{C_1} + \frac{G_1 G_2}{C_1 C_2}} = \frac{\frac{1}{R_1 R_2 C_1 C_2}}{s^2 + s \frac{R_2+R_1}{R_1 R_2 C_1 C_2} + \frac{1}{R_1 R_2 C_1 C_2}}}
$$
**Pasa bajos**
## Vuelvo a la funcion de trasnferencia de orden tres 
Partimos del denominador de la función de transferencia:

$$
s^3 + 6s^2 + 15s + 15
$$

Las raíces del polinomio son:

$$
s_1 = -2.32, \quad s_2 = -1.84 + 1.75j, \quad s_3 = -1.84 - 1.75j \implies (s + 2.32)(s + 1.84 - 1.75j)(s + 1.84 + 1.75j)
$$

$$
(s + 1.84 - 1.75j)(s + 1.84 + 1.75j) = s^2 + s \cdot 2 \cdot 1.84 + 1.84^2 -1.75 j^2 
$$
Finalmente, el polinomio factorizado queda:

$$
s^3 + 6s^2 + 15s + 15 = (s + 2.32)(s^2 + 3.68s + 6.4481)
$$

Por lo tanto, la función de transferencia es:

$$
T(s) = \frac{15}{(s + 2.32)(s^2 + 3.68s + 6.4481)}
$$


## Objetivo

Sintetizar esta función de transferencia **normalizada** utilizando:

- Una estructura **Sallen-Key** con ganancia $ K = 1 $ 
- Un bloque adicional **RC** para el polo real

Una celda Sallen-Key con $K = 1 $ implementa una función de segundo orden:

$$
H_{SK}(s) = \frac{\omega_0^2}{s^2 + \frac{\omega_0}{Q}s + \omega_0^2} =  \frac{\frac{1}{R_1 R_2 C_1 C_2}}{s^2 + s \frac{R_1 + R_2}{R_1 R_2 C_1} + \frac{1}{R_1 R_2 C_1 C_2}}
$$

Esto coincide con el **factor cuadrático** de la función $ T(s) $

La función de transferencia original es de **orden 3**, y ya tenemos un término de segundo orden implementado con Sallen-Key.

El término restante es:

$$
H_{RC}(s) = \frac{1}{s + 2.32}
$$

Este puede ser implementado con un filtro pasabajos simple RC, cuya forma general es:

$$
H_{RC}(s) = \frac{\frac{1}{RC}}{s +  \frac{1}{RC}}
$$

## Conclusión

La función $T(s)$  se puede realizar como:

$$
\boxed{T(s) =  H_{SK}(s) \cdot H_{RC}(s) = \frac{\frac{1}{R_1 R_2 C_1 C_2}}{s^2 + s \frac{R_1 + R_2}{R_1 R_2 C_1} + \frac{1}{R_1 R_2 C_1 C_2}} \cdot  \frac{\frac{1}{RC}}{s +  \frac{1}{RC}}}
$$

Donde:

- $ H_{RC}(s) = \frac{1}{s + 2.32} $ → se implementa con un **filtro pasabajos RC**
- $ H_{SK}(s) = \frac{1}{s^2 + 3.68s + 6.4481} $ → se implementa con una **celda Sallen-Key** con $ K = 1 $

De este modo se logra una **realización en cascada** utilizando bloques estándares de segundo y primer orden.
