# **ENCONTRAR LA FUNCION DE TRANSFERENCIA** #
## ECUACIONES: ##
 $ Vi\cdot G_1= (0-V_A)\cdot S \cdot C + (0-V_A) \cdot G_2+(0-V_O) \cdot G_3$

 $ (V_A-0) \cdot G_3=(0-V_B) \cdot S \cdot C $

 $ (V_B-0) \cdot G_4=(0-V_O) \cdot G_4 $

# 1. $ Vi\cdot G_1= V_A\cdot (S \cdot C+G_2)-V_O \cdot G_3 $ #
# 2. $ V_A \cdot G_3=-V_B \cdot S \cdot C $ #
# 3. $ V_B=-V_O $ #
# Meto $ 3 $ en $ 2 $ : 
$ V_A \cdot G_3=V_O \cdot S \cdot C  \rightarrow\ V_A= \frac{V_O \cdot S \cdot C}{G_3}$
# Ahora en $ 1 $ :
# $ Vi\cdot G_1= \cdot \frac{V_O \cdot S \cdot C}{G_3}(S \cdot C+G_2)-V_O \cdot G_3 = -V_O \cdot (\frac{S^2 \cdot C^2+ S \cdot C \cdot G_2}{G_3}+G_3)= -V_O \cdot (\frac{S^2 \cdot C^2+ S \cdot C \cdot G_2}{G_3^2}) $ #
# $ V_i=-V_O \cdot ( \frac{S^2 \cdot C^2+ S \cdot C \cdot G_2}{G_3^2}) \cdot \frac{1}{G_1}  $ #
## $ \frac{V_O}{V_i}=-\frac{G_3 \dot G_1}{S^2 \cdot C^2+ S\frac{g_2}{c} + G_3^2} $
## Simplifico y obtengo $H(s)$:
##  $$ \boxed{H(S)=\frac{V_O}{V_i}=-\frac{\frac{G_3 \dot G_1}{C^2}}{S^2+ S \cdot \frac{G_2}{C} + \frac{G_3^2}{C^2}}= -\frac{\frac{1}{R_3 \dot R_1 \dot C^2}}{S^2+ S \cdot \frac{1}{R_2 \dot C} + \frac{1}{R_3^2 \dot C^2}}}  $$  ## 

**Pasa bajos de segundo orden**

---

# Multiplico y divido el numerador por $G_3$
## $ \frac{V_O}{V_i}=-\frac{\frac{G_3 \dot G_1}{C^2} \cdot \frac{G_3}{G_3} }{S^2+ S \cdot \frac{G_2}{C} + \frac{G_3^2}{C^2}} =\frac{-\frac{G_1}{G_3} \cdot \frac{G_3^2}{C^2} }{S^2+ S \cdot \frac{G_2}{C} + \frac{G_3^2}{C^2}} $

---

### Parametros

- En **conductancias**:

$$K = -\frac{G_1}{G_3} \quad , \quad \omega_0 = \frac{G_3}{C} \quad , \quad Q = \frac{G_3}{G_2}$$

- En **resistencias**:

$$K = -\frac{R_3}{R_1} \quad , \quad \omega_0 = \frac{1}{R_3 C} \quad , \quad Q = \frac{R_2}{R_3}$$

---


### Normalizacin

Para normalizar:
- Elijo $G_3$ porque aparece en los tres parametros.

$Z_0= G_3  \rightarrow\ Z_N=\frac{Z}{Z_0}=1 \rightarrow\ G_3=1$
Por lo tanto:

- La frecuencia angular normalizada es:

## $\omega_0 = \frac{G_3}{C}  = 1 \rightarrow\  C=1 $  (Esto en realidad es $C=\frac{1}{R_3})$ # 

- La ganancia:

## $K = -G_1=-\frac{1}{R_1}$ #

- El factor de selectividad:

## $Q = \frac{1}{G_2}=R_2$ #

As, la funcin de transferencia normalizada se expresa:

## $$\boxed{H(s) = \frac{-G_1}{s^2 + s G_2 + 1}}$$ ##

donde los parmetros libres son $ G_1 $ y $ G_2 $.

## Sistema de ecuaciones

Partimos del siguiente sistema:
# 1. $ V_i G_1 = (0 - V_o) sC + (0 - V_o) G_2 + (0 - V_a) G_3 $ #
# 2. $ (V_o - 0) G_3 = (0 - V_b) sC $ #
# 3. $ (V_b - 0) G_4 = (0 - V_a) G_4 $ #

## Simplificaciones

De la ecuacin $3$ obtenemos:

$V_b = -V_a$

Reemplazando en $2$:

$(V_o) G_3 = (-(-V_a)) sC$

$V_o G_3 = V_a sC$

Despejando $ V_a $:

$V_a = \frac{V_o G_3}{sC}$

## Reemplazamos $ V_a $ en la ecuacin $ 1 $:

$V_i G_1 = -V_o sC - V_o G_2 - V_a G_3$

$V_i G_1 = -V_o sC - V_o G_2 - \left( \frac{V_o G_3}{sC} \right) G_3$

$V_i G_1 = -V_o \left( sC + G_2 + \frac{G_3^2}{sC} \right)$

Multiplicamos todo por $sC$ para eliminar el denominador:

$V_i G_1 \cdot sC = -V_o \left( (sC + G_2) sC + G_3^2 \right)$

Expandimos:

$V_i G_1 sC = -V_o \left( s^2 C^2 + sC G_2 + G_3^2 \right)$

$V_o = -\frac{V_i G_1 sC}{s^2 C^2 + sC G_2 + G_3^2}  \rightarrow\  \frac{V_o}{V_i} = -\frac{G_1 sC}{s^2 C^2 + sC G_2 + G_3^2}$

## Dividimos numerador y denominador por $ C^2 $:

$$\boxed{H(s) = -\frac{G_1}{C} \cdot \frac{s}{s^2 + \frac{G_2}{C} s + \frac{G_3^2}{C^2}}=-\frac{1}{R_1 C} \cdot \frac{s}{s^2 + \frac{1}{R_2 C} s + \frac{1}{R_3^2 C^2} }}$$
**Pasa banda de segundo orden**

---

## Definimos los parmetros:

- $$ K = -\frac{G_1}{C} \quad \rightarrow \quad K = -\frac{1}{R_1 C} $$
- $$ \omega_0 = \frac{G_3}{C} \quad \rightarrow \quad \omega_0 = \frac{1}{R_3 C} $$
- $$ Q = \frac{G_3}{G_2} \quad \rightarrow \quad Q = \frac{R_2}{R_3} $$

---

### Normalizacin (IDEM PASA BAJOS)

Para normalizar:
- Elijo $G_3$ porque aparece en los tres parametros.

$Z_0= G_3  \rightarrow\ Z_N=\frac{Z}{Z_0}=1 \rightarrow\ G_3=1$
Por lo tanto:

- La frecuencia angular normalizada es:

## $\omega_0 = \frac{G_3}{C}  = 1 \rightarrow\  C=1 $  (Esto en realidad es $C=\frac{1}{R_3})$ # 

- La ganancia:

## $K = -G_1=-\frac{1}{R_1}$ #

- El factor de selectividad:

## $Q = \frac{1}{G_2}=R_2$ #
As, la funcin de transferencia normalizada se expresa:

## $$\boxed{H(s) = G_1 \cdot \frac{s}{s^2 + s G_2 + 1} = \frac{1}{R_1} \cdot \frac{s}{s^2 + \frac{s}{R_2} + 1}}$$

```python
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import bodePlot, pzmap, GroupDelay
# PASA BAJOS
# Parmetros normalizados 
R1 = 2  # Valor arbitrario
R2 = 3  # Valor arbitrario
R3 = 1    # ohm (por normalizacin)

K = -R3 / R1
Q = R2 / R3
w0 = 1  # normalizado

# Funcin de transferencia normalizada
num = [0,0, K * (w0**2)]            # numerador: K * w0**2 * s
den = [1, w0/Q, w0**2]          # denominador: s^2 + (w0/Q) * s + w0^2

H = TransferFunction(num, den)

# Plots
bodePlot(H, fig_id=1, filter_description='H(s) normalizado')
pzmap(H, fig_id=2, filter_description='H(s) normalizado')
GroupDelay(H, fig_id=3, filter_description='H(s) normalizado')

plt.show()
```
