##  Cálculo del parámetro $\varepsilon$

Recordando que la atenuación en la banda pasante se relaciona con $\varepsilon$ mediante:

$$
\alpha_{\text{max}} = -10 \log_{10} \left( \frac{1}{1 + \varepsilon^2} \right)
$$

Despejando:

$$
10^{\frac{\alpha_{\text{max}}}{10}} = 1 + \varepsilon^2 \\
\Rightarrow \varepsilon^2 = 10^{\frac{1}{10}} - 1 = 10^{0.125} - 1 \approx 0.2589
$$

$$ \boxed{\varepsilon \approx 0.5088}$$
**No se trata de un Butterworth ya que $\varepsilon \neq 1$**

---

##  Cálculo del orden $n$ del filtro
$$\alpha_{\text{min}}=10 \log_{10} (\varepsilon^2 \omega_s^{2n}+1)$$

$
10^{\frac{\alpha_{\text{min}}}{10}} = \varepsilon^2 \omega_s^{2n} + 1
$

$
\omega_s^{2n} = \frac{10^{\frac{\alpha_{\text{min}}}{10}} - 1}{\varepsilon^2}
$

$
\log_{10}(\omega_s^{2n}) = \log_{10} \left( \frac{10^{\frac{\alpha_{\text{min}}}{10}} - 1}{\varepsilon^2} \right)
$

$
2n \log_{10}(\omega_s) = \log_{10} \left( \frac{10^{\frac{\alpha_{\text{min}}}{10}} - 1}{\varepsilon^2} \right)
$

$$
n = \frac{1}{2} \cdot \frac{ \log_{10} \left( \frac{10^{\frac{\alpha_{\text{min}}}{10}} - 1}{\varepsilon^2} \right)}{ \log_{10}(\omega_s) }
$$

Sabemos que:

- $\alpha_{\text{min}} = 12$ dB
- $\alpha_{\text{max}} = 1$ dB
- $\omega_s = 2$
- $\varepsilon^2 = \approx 0.2589$

Reemplazamos:

$
n = \frac{1}{2} \cdot \frac{ \log_{10} \left( \frac{10^{\frac{12}{10}} - 1}{0.2589} \right)}{ \log_{10}(2) } \approx 2.92 \Rightarrow  \boxed{n = 3}
$

##  Forma de la función de transferencia

$$T(s) \cdot T(-s) = \frac{1}{Q(s) Q(-s)} = \frac{1}{1 + \varepsilon^2 \omega^{2n}} = \frac{1}{1 + \varepsilon^2 \omega^{2 \cdot 3}} =  \frac{1}{1 + \varepsilon^2 \omega^{6}} = \frac{\frac{1}{\varepsilon^2}}{\frac{1}{\varepsilon^2}+\omega^{6}}$$
 $$s=j \cdot \omega  \Rightarrow  T(s) \cdot T(-s) = \frac{\frac{1}{\varepsilon^2}}{\frac{1}{\varepsilon^2} - s^6}$$

 $$
\frac{\frac{1}{\varepsilon^2}}{\frac{1}{\varepsilon^2} - s^6} =
\left\{
\begin{array}{ll}
T(s) &= \dfrac{a_0}{a_3 s^3 + a_2 s^2 + a_1 s + a_0} \\
T(-s) &= \dfrac{a_0}{-a_3 s^3 + a_2 s^2 - a_1 s + a_0}
\end{array}
\right.
$$

$$\frac{1}{\varepsilon^2} - s^6 = (a_3 s^3 + a_2 s^2 + a_1 s + a_0) \cdot (-a_3 s^3 + a_2 s^2 - a_1 s + a_0)$$

$$
\begin{align*}

&\quad -a_1 a_3 s^4 + a_1 a_2 s^3 - a_1^2 s^2 + a_1 a_0 s \\
&\quad - a_0 a_3 s^3 + a_0 a_2 s^2 - a_0 a_1 s + a_0^2
\end{align*}
$$

$$
\begin{align*}
\text{Grado } s^6: & \quad -a_3^2 = -1 \Rightarrow a_3^2 = 1 \\
\text{Grado } s^5: & \quad a_3 a_2 - a_2 a_3 = 0 \\
\text{Grado } s^4: & \quad -a_3 a_1 - a_1 a_3 + a_2^2 = -2 a_1 a_3 + a_2^2 = 0 \\
\text{Grado } s^3: & \quad a_3 a_0 - a_2 a_1 + a_1 a_2 - a_0 a_3 = (a_3 a_0 - a_0 a_3) + 0 = 0 \\

\text{Grado } s^1: & \quad a_1 a_0 - a_0 a_1 = 0 \\
\text{Grado } s^0: & \quad a_0^2 = \frac{1}{\varepsilon^2}
\end{align*}
$$


Entonces obtenemos el sistema:

## Funcion de transferencia .

Partimos del siguiente sistema de ecuaciones:

1. $$ (V_i - V_b) g_{m1} = s C_1 V_a $$
2. $$ (V_a - V_b) g_{m2} = s C_2 V_b $$
3. $$ \frac{V_b - V_o}{R} = s C_3 V_o $$

---

## Paso 1: Despejar $V_b$ de la ecuación (3)

$$
V_b - V_o = R s C_3 V_o \Rightarrow V_b = V_o (1 + R s C_3)
$$

---

## Paso 2: Despejar $V_b $ de la ecuación (1)

De (1):

$$
(V_i - V_b) g_{m1} = s C_1 V_a
\Rightarrow V_i g_{m1} - V_b g_{m1} = s C_1 V_a
$$

Despejamos $ V_b $:

$$
V_b = \frac{V_i g_{m1} - s C_1 V_a}{g_{m1}}
$$

---

## Paso 3: Despejar $ V_a $ de la ecuación (2)

De (2):

$$
(V_a - V_b) g_{m2} = s C_2 V_b
\Rightarrow V_a g_{m2} - V_b g_{m2} = s C_2 V_b
$$

Despejamos $ V_a $:

$$
V_a = \frac{g_{m2} + s C_2}{g_{m2}} V_b
$$

---

## Paso 4: Sustituir $ V_a $ en la expresión de $ V_b $ del paso 2

Recordemos:

$$
V_b = \frac{V_i g_{m1} - s C_1 V_a}{g_{m1}}
$$

Sustituimos $ V_a $:

$$
V_b = \frac{V_i g_{m1} - s C_1 \cdot \frac{g_{m2} + s C_2}{g_{m2}} V_b }{g_{m1}}
$$

Multiplicamos y reordenamos:

$$
V_b = \frac{V_i g_{m1} - \frac{s C_1 (g_{m2} + s C_2)}{g_{m2}} V_b }{g_{m1}}
$$

Multiplicamos ambos lados por $g_{m1} $:

$$
g_{m1} V_b = V_i g_{m1} - \frac{s C_1 (g_{m2} + s C_2)}{g_{m2}} V_b
$$

Pasamos términos de $ V_b $ al mismo lado:

$$
\left( g_{m1} + \frac{s C_1 (g_{m2} + s C_2)}{g_{m2}} \right) V_b = V_i g_{m1}
$$

Multiplicamos el denominador para eliminar fracciones internas:

Denominador común: $ g_{m2} $

$$
V_b = \frac{V_i g_{m1} g_{m2}}{g_{m1} g_{m2} + s C_1 g_{m2} + s^2 C_1 C_2}
$$

---

## Paso 5: Sustituir $V_b $ en la ecuación (3) para obtener $V_o $

Recordamos que:

$$
V_b = V_o (1 + R s C_3)
\Rightarrow V_o = \frac{V_b}{1 + R s C_3}
$$

Entonces:

$$
V_o = \frac{1}{1 + R s C_3} \cdot \frac{V_i g_{m1} g_{m2}}{g_{m1} g_{m2} + s C_1 g_{m2} + s^2 C_1 C_2}
$$

---

## Paso 6: Función de transferencia $T(s) = \frac{V_o}{V_i} $

Multiplicamos los denominadores:

$$
T(s) = \frac{g_{m1} g_{m2}}{
s^3 C_1 C_2 R C_3
+ s^2 (C_1 C_2 + C_1 g_{m2} R C_3)
+ s (C_1 g_{m2} + g_{m1} g_{m2} R C_3)
+ g_{m1} g_{m2}
}
$$

Sacamos como denominador común a  $C_1 C_2 C_3 R$:

$$
\boxed{
T(s) = \frac{g_{m1} g_{m2}}{C_1 C_2 C_3 R} \cdot
\frac{1}{
s^3
+ s^2 \left( \frac{C_2 + g_{m2} R C_3}{C_2 R C_3} \right)
+ s \cdot \left( \frac{g_{m2}(C_3 g_{m1} + C_1)}{C_1 C_2 C_3 R} \right)
+ \frac{g_{m1} g_{m2}}{C_1 C_2 R C_3}
}}
$$
 Para normalizarla la voy a partir en 2:
Teniamos
$$ V_o = \frac{1}{1 + R s C_3} \cdot \frac{V_i g_{m1} g_{m2}}{g_{m1} g_{m2} + s C_1 g_{m2} + s^2 C_1 C_2} \Rightarrow \frac{V_o}{V_i} = \frac{1}{1 + R s C_3} \cdot \frac{ g_{m1} g_{m2}}{g_{m1} g_{m2} + s C_1 g_{m2} + s^2 C_1 C_2}$$

$$
T_1 = \frac{1}{1 + R s C_3} = \frac{\frac{1}{R C_3}}{\frac{1}{R C_3} + s} = \frac{\omega_0^\prime}{s + \omega_0^\prime}
$$
$$
T_2 = \frac{ g_{m1} g_{m2}}{g_{m1} g_{m2} + s C_1 g_{m2} + s^2 C_1 C_2} = \frac{\frac{ g_{m1} g_{m2}}{C_1 C_2}}{\frac{g_{m1} g_{m2}}{C_1 C_2} +(\frac{ g_{m2}}{ C_2}s) + s^2} = \frac{\omega_0^2}{s^2 + s (\frac{\omega_0 }{ Q})+ \omega_0^2 }
$$

$$
T_1(s) \cdot T_2(s) = \frac{\omega_0'}{s + \omega_0'} \cdot \frac{\omega_0^2}{s^2 + \frac{\omega_0}{Q}s + \omega_0^2}
$$

Tambien teniamos: 
$$T(s)= \frac{1.96}{s^3 + 2.50s^2 + 3.14s + 1.96}$$

WolframAlpha nos da las siguientes raíces del polinomio cúbico:

- $ s_1 = -1.24 $
- $ s_2 = -0.63 - 1.09i $
- $ s_3 = -0.63 + 1.09i $

Entonces, el polinomio se puede escribir como el producto:

$$
P(s) = (s - s_1)(s - s_2)(s - s_3)
$$

Reemplazando las raíces:

$$
P(s) = (s + 1.24)\left[(s + 0.63 + 1.09i)(s + 0.63 - 1.09i)\right]
$$

$$
(s + 0.63 + 1.09i)(s + 0.63 - 1.09i) = (s + 0.63)^2 + (1.09)^2 = = s^2 + 1.26s + 1.585
$$

Redondeando:

$$
P(s) = (s + 1.24)(s^2 + 1.26s + 1.59)
$$

Esta es la factorización del polinomio original:

$$
s^3 + 2.5s^2 + 3.14s + 1.96 = (s + 1.24)(s^2 + 1.26s + 1.59)
$$

Volvemos a la exprecion original:
$$T(s)= \frac{1.96}{ (s + 1.24)(s^2 + 1.26s + 1.59)}$$


$$
\frac{\omega_0^\prime}{s + \omega_0^\prime} \cdot \frac{\omega_0^2}{s^2 + \frac{\omega_0}{Q}s + \omega_0^2} = \frac{1.96}{(s + 1.24)(s^2 + 1.26s + 1.59)}
$$

- $ \omega_0' = 1.24 $
- $ \omega_0^2 = 1.59 \Rightarrow  \omega_0 \approx 1.26$
- $ \frac{\omega_0}{Q} = 1.26  \Rightarrow Q  \approx 1 $
- Tomando $ C_1 = C_2 = C_3 = 1 $

Sabemos que:

$$
\omega_0' = \frac{1}{RC_3} \Rightarrow  R = \frac{1}{\omega_0'} = \frac{1}{1.24} \approx 0.806
$$
$$
\omega_0^2 = \frac{g_{m1} g_{m2}}{C_1 C_2} \Rightarrow g_{m1} g_{m2} = \omega_0^2 = 1.59
$$
$$
\frac{\omega_0}{Q} = \frac{C_2}{g_{m2}} \Rightarrow g_{m2} = \frac{Q}{\omega_0} = 1.26
$$

A partir del producto:

$$
g_{m1} = \frac{g_{m1} g_{m2}}{g_{m2}} = \frac{1.59}{1.26} \approx 1.26
$$

Los valores que satisfacen la función de transferencia deseada, tomando $ C_1 = C_2 = C_3 = 1 $, son:

$$
\boxed{
R \approx 0.806 \quad,\quad g_{m1} \approx 1.26 \quad,\quad g_{m2} \approx 1.26
}
$$
