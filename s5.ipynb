## Obtener la función transferencia normalizada del prototipo pasa bajo que satisfaga el requerimiento del filtro pasa banda.
$$ j\Omega = \frac{ (j\omega)^2 + \omega_0^2 }{ j\omega \cdot \frac{\omega_0}{Q} } \Rightarrow \Omega = \frac{ Q ( \omega^2 - \omega_0^2 ) }{ \omega \omega_0 } $$

Aclaracion: $\Omega$ pertenece al pasa bajos y $\omega$ al pasa banda.

Si $\omega_0 = 1$ :
$$\Omega = \frac{ Q ( \omega^2 - 1 ) }{ \omega } \Rightarrow
\left\{
\begin{array}{ll}
\Omega_1(\omega=\omega_{s1}) = 2.64\\
\Omega_2(\omega=\omega_{s2}) = 5.15\\
\end{array}
\right.$$

Averiguo el orden del filtro:

$$\alpha_{MIN} = 10log(1+\epsilon^2 \cdot C_n^2(\Omega_s)) \quad \land \quad \epsilon^2=10^{\frac{\alpha_{MAX}}{10}}-1 \Rightarrow C_n(\Omega_s)= \sqrt{\frac{10^{\frac{\alpha_{MIN}}{10}}-1}{10^{\frac{\alpha_{MAX}}{10}}-1}}$$

¿Que datos uso?

Se que $\alpha_{MAX}=0.5dB$ y que tengo 2 $\alpha_{MIN}$ que dependen del valor de $\Omega_s$:

Ademas tengo:
$$C_0=1 \quad C_1=\omega \quad C_2=2\cdot \omega^2 - 1 \quad C_3=4 \cdot \omega^3-3 \cdot \omega$$



Evaluo:$$\alpha_{MIN}=16dB \Rightarrow \Omega_1 = 2.64$$

$$\sqrt{\frac{10^{\frac{16}{10}}-1}{10^{\frac{0.5}{10}}-1}} \approx 17.83$$
$n=2 \quad \land \quad C_2=2\cdot \omega^2 - 1$
$$\sqrt{\frac{10^{\frac{16}{10}}-1}{10^{\frac{0.5}{10}}-1}} \le 2\cdot \Omega_1^2 - 1 \Rightarrow 17.83 \le 12.93$$
**No cumple**

$n=3 \quad \land \quad C_3=4 \cdot \omega^3-3 \cdot \omega$
$$\sqrt{\frac{10^{\frac{16}{10}}-1}{10^{\frac{0.5}{10}}-1}} \le 4 \cdot \omega^3-3 \cdot \omega \Rightarrow 17.83 \le 65.67$$
**Cumple**

$$\boxed{\alpha_{MIN}=16dB \quad \land \quad \Omega_1 = 2.64 \quad \land \quad n=3}$$

Evaluo $$\alpha_{MIN}=24dB \Rightarrow \Omega_2 = 5.15$$
$n=2 \quad \land \quad C_2=2\cdot \omega^2 - 1$
$$\sqrt{\frac{10^{\frac{24}{10}}-1}{10^{\frac{0.5}{10}}-1}} \le 2\cdot \Omega_2^2 - 1 \Rightarrow 45.28 \le 52.04$$
**Cumple**
$$\boxed{\alpha_{MIN}=24dB \quad \land \quad \Omega_2 = 5.15 \quad \land \quad n=2}$$

El mas exigente es el de orden 3, por lo que el **pasa bajos que satisfase el requerimento del pasa bajos es**:
$$\boxed{T_{LP}=\frac{1}{s+1} \cdot \frac{1}{s^2+s\frac{1}{Q}+1}}$$

## Aplicar el núcleo de transformación pasa banda y obtener la transferencia
$$\epsilon^2=10^{\frac{\alpha_{MAX}}{10}}-1=10^{\frac{0.5}{10}}-1=0.122 \Rightarrow \epsilon=0.349$$

$$|T(s)|^2=T(s).T(-s)=\frac{1}{1+\epsilon^2.C_n^2(\omega)} = \frac{1}{1+\epsilon^2.C_3^2(\omega)}$$

$$\frac{1}{1+\epsilon^2.(4\omega^3-3\omega)^2} = \frac{\frac{1}{\epsilon^2}}{16.\omega^6-24.\omega^4+9.\omega^2+\frac{1}{\epsilon^2}} = \frac{\frac{1}{16.\epsilon^2}}{\omega^6-\frac{3}{2}.\omega^4+\frac{9}{16}.\omega^2+\frac{1}{16.\epsilon^2}}$$

 $$
\omega = \frac{s}{j} \Rightarrow \frac{\frac{1}{16.\epsilon^2}}{-s^6-\frac{3}{2}.s^4-\frac{9}{16}.s^2+\frac{1}{16.\epsilon^2}} =
\left\{
\begin{array}{ll}
T(s) &= \dfrac{a_0}{ s^3 + a_2 s^2 + a_1 s + a_0} \\
T(-s) &= \dfrac{a_0}{- s^3 + a_2 s^2 - a_1 s + a_0}
\end{array}
\right.
$$

$$
\begin{align*}
\text{Grado } s^6: & \quad -1 = -1 \\
\\
\text{Grado } s^5: & \quad a_2 - a_2 = 0 \\
\\
\text{Grado } s^4: & \quad -a_1 - a_1 + a_2^2 = -2a_1 + a_2^2 = -\frac{3}{2} \\
\\
\text{Grado } s^3: & \quad a_0 - a_2 a_1 + a_1 a_2 - a_0 = 0 \\
\\
\text{Grado } s^2: & \quad a_2 a_0 - a_1^2 + a_0 a_2 = 2 a_0 a_2 - a_1^2 = -\frac{9}{16} \\
\\
\text{Grado } s^1: & \quad a_1 a_0 - a_0 a_1 = 0 \\
\\
\text{Grado } s^0: & \quad a_0^2 = \frac{1}{16 \epsilon^2}
\end{align*}
$$

Del grado 4 saco: 

$$a_1=\frac{1}{2}.a_2^2+\frac{3}{4}$$

Del grado 2 obtengo:

$$2.a_2.a_0-(\frac{1}{2}.a_2^2+\frac{3}{4})^2+\frac{9}{16} = 0 \Rightarrow 2.a_2.a_0-\frac{a_2^4}{4}-\frac{3}{4}a_2^2=0$$

Resolviendo queda:
$$\boxed{a_2 = 1.253 \quad \land \quad a_1 = 1.535 \quad \land \quad a_0 = 0.716}$$
La expresion queda:
$$T_{LP}(s)=\frac{0.716}{s^3+1.253s^2+1.535s+0.716}$$

## Realizar la simulación numérica en python. Graficar la respuesta en frecuencia del filtro pasa banda total, y las de las SOS. 

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy import signal as sig
from scipy.signal import TransferFunction

from pytc2.sistemas_lineales import (
    analyze_sys,
    pretty_print_bicuad_omegayq,
    tf2sos_analog,
    pretty_print_SOS,
    pretty_print_lti
)

from IPython.display import display, Markdown

# --- Parámetros del diseño --- #

ripple = 0.5             # Ripple en dB 
n = 3                    # Orden del pasa bajos
Q = 5                    # Q deseado para la transformación pasa banda
bw = 1 / Q               # Ancho de banda de la transformación pasa banda
e = np.sqrt(10**(ripple / 10) - 1)  # Factor de ripple

# --- Diseño del filtro LP Chebyshev --- #

z, p, k = sig.cheb1ap(n, ripple)  # Diseño analógico pasa bajos (normalizado)

display(Markdown("## Filtro pasa bajos Chebyshev de orden 3"))

# --- Transformación a forma de función de transferencia --- #
num_lp, den_lp = sig.zpk2tf(z, p, k)

# Conversión a SOS (secciones de segundo orden)
sos_lp = tf2sos_analog(num_lp, den_lp)

pretty_print_SOS(sos_lp, mode='omegayq')

# ---Transformación LP → BP --- #

display(Markdown("## Filtro pasa banda de orden 6 (via LP → BP)"))

# --- Transformación pasa bajos → pasa banda --- #
num_bp, den_bp = sig.lp2bp(num_lp, den_lp, bw=bw)

# --- Conversión a SOS --- #
sos_bp = tf2sos_analog(num_bp, den_bp)

display(Markdown("### Forma SOS (Secciones de segundo orden)"))
pretty_print_SOS(sos_bp, mode='omegayq')

display(Markdown("### Forma estándar (no SOS)"))
pretty_print_lti(num_bp, den_bp)

# --- Análisis del sistema --- #

display(Markdown("## Análisis del filtro pasa banda"))

# --- Análisis gráfico: módulo, fase, polos y ceros --- #
analyze_sys(sig.TransferFunction(num_bp, den_bp), same_figs=False)
plt.show()
```
