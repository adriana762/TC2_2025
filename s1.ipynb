## El objetivo es encontrar la impedancia de entrada:
$Z_{in}(s) = \frac{V_x}{I_x}$

---
## Por nodos

**1.** $V_x (Y_2 + Y_3) - V_A \cdot Y_2 - V_2 \cdot Y_3 = 0$

**2.** $V_x (G_4 + G_5) - V_2 \cdot G_4 = 0$

**3.** $(V_x - V_A) \cdot Y_1 = I_x$

- $I_x $: Corriente de entrada al sistema
  
---

## Desarrollo

### Partimos de la ecuación 2:

$V_x (G_4 + G_5) - V_2 \cdot G_4 = 0$

Despejamos $V_x $:
$V_x = \frac{G_4}{G_4 + G_5} \cdot V_2$

Y también se puede despejar $V_2$:
$V_2 = \frac{G_4 + G_5}{G_4} \cdot V_x$

---

### Reemplazamos en la ecuación 1:

$V_x (Y_2 + Y_3) - V_A \cdot Y_2 - V_2 \cdot Y_3 = 0$

Sustituyendo \( V_2 \):
$V_x (Y_2 + Y_3) - V_A \cdot Y_2 - \left( \frac{G_4 + G_5}{G_4} \cdot V_x \right) Y_3 = 0$

Agrupamos los términos con $ V_x $:
$V_A \cdot Y_2 = V_x \left[ (Y_2 + Y_3) - \left( \frac{G_4 + G_5}{G_4} \cdot Y_3 \right) \right]$

Finalmente:
$V_A = V_x \cdot \left( \frac{(Y_2 + Y_3) - \left( \frac{G_4 + G_5}{G_4} \cdot Y_3 \right)}{Y_2} \right)$

---

## Sustituyo en la ecuación 3

$(V_x - V_A) \cdot Y_1 = I_x$

Ya tenemos:

$V_A = V_x \cdot \left( \frac{(Y_2 + Y_3) - \left( \frac{G_4 + G_5}{G_4} \cdot Y_3 \right)}{Y_2} \right)$

Sustituimos en la ecuación:

$\left( V_x - V_x \cdot \left( \frac{(Y_2 + Y_3) - \left( \frac{G_4 + G_5}{G_4} \cdot Y_3 \right)}{Y_2} \right) \right) \cdot Y_1 = I_x$

Factorizamos $V_x$:

$V_x \cdot \left[ 1 - \left( \frac{(Y_2 + Y_3) - \left( \frac{G_4 + G_5}{G_4} \cdot Y_3 \right)}{Y_2} \right) \right] \cdot Y_1 = I_x$

---

##  Finalmente: impedancia de entrada

$Z_{in}(s) = \frac{V_x}{I_x} = \frac{1}{Y_1 \cdot \left[ 1 - \left( \frac{(Y_2 + Y_3) - \left( \frac{G_4 + G_5}{G_4} \cdot Y_3 \right)}{Y_2} \right) \right]}$

## Reduciendo la expresión:

## $$\boxed{Z_{in}(s) = \frac{Y_2 \cdot G_4}{Y_1 \cdot Y_3 \cdot G_5}}$$

---

## Queremos que $Z_{in}(s)$ responda como un inductor unitario

## entonces:  $Z_{in}(s) = sL$, con $L = 1$, por lo tanto:

- La impedancia debe tener la forma $Z_{in}(s) = s$
- Eso implica que debe haber un **capacitor** (para que aparezca la variable $s$ en el numerador)
- El resto de los elementos deben ser **resistencias**

Dado que:

- $Y_2$ debe ser un capacitor: $Y_2 = sC_2$
- $G_4$ representa una resistencia: $G_4 = 1/R_4$
- Y todos los componentes tienen valor 1: $C_2 = R_1 = R_3 = R_4 = R_5 = 1$

Entonces:

- $Y_1 = 1/R_1 = 1$
- $Y_3 = 1/R_3 = 1$
- $G_5 = 1/R_5 = 1$

# Finalmente, reemplazando en la expresión:

## $$\boxed{Z_{in}(s) = \frac{Y_2 \cdot G_4}{Y_1 \cdot Y_3 \cdot G_5} = \frac{sC_2 \cdot R_1 \cdot R_3 \cdot R_5}{R_2} = s} $$


Esto corresponde a una **impedancia inductiva de valor unitario**, es decir, un **inductor ideal de 1 H**.

## Sistema de ecuaciones para obtener $H(s) = \frac{V_2(s)}{V_1(s)}$

Las siguientes ecuaciones surgen del análisis nodal del circuito:

**1.** $V_x \left( sC_2 + \frac{1}{R_3} \right) - V_A \cdot sC_2 - \frac{V_2}{R_3} = 0$    
**2.** $V_x \left( \frac{1}{R_4} + \frac{1}{R_5} \right) - \frac{V_2}{R_4} = 0$   
**3.** $V_x \left( \frac{1}{R} + sC + \frac{1}{R_1} \right) - \frac{V_1}{R} - \frac{V_A}{R_1} = 0$
   
##  Sustituciones 

De la ecuación **2** despejamos $V_x$:

$V_x = \frac{V_2 \cdot R_5}{R_4 + R_5}$

De la ecuación **3** despejamos $V_A$:

$V_A =  \frac{V_2 \cdot R_5}{R_4 + R_5} \cdot \frac{R_1 + R_1 R s C + R}{R} - \frac{V_1 R_1}{R}$

Y reemplazando esto en la ecuacion **1** obtenemos:

$\frac{V_2 \cdot R_5}{R_4 + R_5} \left( sC_2 + \frac{1}{R_3} \right) - (\frac{V_2 \cdot R_5}{R_4 + R_5} \cdot \frac{R_1 + R_1 R s C + R}{R} - \frac{V_1 R_1}{R}) \cdot sC_2 - \frac{V_2}{R_3} = 0$ 

$\frac{V_2 \cdot R_5}{R_4 + R_5}\cdot sC_2 + \frac{V_2 \cdot R_5}{R_4 + R_5} \cdot \frac{1}{R_3} - \frac{V_2 \cdot R_5}{R_4 + R_5} \cdot \frac{R_1 + R_1 R s C + R}{R} \cdot sC_2 + \frac{V_1 R_1}{R} \cdot sC_2 - \frac{V_2}{R_3} = 0$

$\frac{V_1 R_1}{R} \cdot sC_2=-\frac{V_2 \cdot R_5}{R_4 + R_5}\cdot sC_2 - \frac{V_2 \cdot R_5}{R_4 + R_5} \cdot \frac{1}{R_3} + \frac{V_2 \cdot R_5}{R_4 + R_5} \cdot \frac{R_1 + R_1 R s C + R}{R} \cdot sC_2 + \frac{V_2}{R_3}$

$V_1 \cdot \frac{R_1}{R} \cdot sC_2=V_2 \cdot (\frac{R_5}{R_4 + R_5} \cdot \frac{R_1 + R_1 R s C + R}{R} \cdot sC_2 + \frac{1}{R_3}-\frac{R_5}{R_4 + R_5}\cdot sC_2 - \frac{R_5}{R_4 + R_5} \cdot \frac{1}{R_3} )$

# Finalmlente 

## $$\boxed{H(s) =  \frac{V_2(s)}{V_1(s)} = \frac{(s \cdot \frac{1}{RC}) \cdot \left( \frac{R_4}{R_5} + 1 \right)}{s^2 + \frac{1}{CR}s + \frac{R_4}{C_2 \cdot C \cdot R_1 \cdot R_3 \cdot R_5}}} $$ ##

## Parametros:
$$K = \left( \frac{R_4}{R_5} + 1 \right) \quad ,  \omega_0 = \sqrt{\frac{R_4}{C_2 \cdot C \cdot R_1 \cdot R_3 \cdot R_5}} \quad ,  Q = C \cdot R$$

**Pasa banda de segundo orden**

---

## Parámetros definidos para la normalización

- Se elige:  
  $$
  C_2 = C = 1
  $$

- También se tiene:  
  $$
  R_5 = R_3 = 1
  $$

- $R$, $R_4$, $R_1$ son **libres**, pero con la condición de que:  
  $$
  R_4 = R_1
  $$

- La **ganancia** queda entonces como:  
  $$
  K = \frac{R_4}{R_5} + 1 = R_4 + 1
  $$

- El **factor de selectividad** es:  
  $$
  Q = R
  $$

---

## Datos de desnormalización

- **Capacitor desnormalizado**:  
  $$
  C = 10\,\text{nF}
  $$

- **Frecuencia de referencia**:  
  $$
  f_0 = 10\,\text{kHz}
  $$

- **Norma de impedancia**:  
  $$
  Z_0 = \frac{1}{2\pi f_0 C} = \frac{1}{2\pi \cdot 10^4 \cdot 10^{-8}} \approx 1591\,\Omega
  $$


  ### Desnormalización de la función de transferencia

Partimos de la función de transferencia **normalizada**:

$$
H(s) = \frac{0.5s}{s^2 + \frac{s}{Q} + 1}
$$

donde:
- $ K = \frac{R_4}{R_5} + 1 = 10 $
- $ Q = R = 20 $

Queremos obtener la función **desnormalizada**, usando los siguientes parámetros:

- Frecuencia de referencia: $ f_0 = 10 \, \text{kHz} $
- $ \omega_0 = 2\pi f_0 = 62831.85 \, \text{rad/s} $
- Capacitor: $ C = 10 \, \text{nF} $
- Norma de impedancia:  
  $$
  Z_0 = \frac{1}{C \cdot \omega_0} = \frac{1}{10^{-8} \cdot 62831.85} \approx 1591 \, \Omega
  $$

El cambio de variable es:

$$
s = \frac{S}{\omega_0}
$$

Sustituyendo en la función normalizada:

$$
H(S) = \frac{0.5 \cdot \frac{S}{\omega_0}}{\left( \frac{S}{\omega_0} \right)^2 + \frac{1}{Q} \cdot \frac{S}{\omega_0} + 1}
$$

Multiplicamos numerador y denominador por $\omega_0^2$ para eliminar los denominadores:

$$
H(S) = \frac{0.5 \cdot S \cdot \omega_0}{S^2 + \frac{\omega_0}{Q} S + \omega_0^2}
$$

---

**Resultado final:**

## $$ \boxed{H(S) = \frac{0.5 \cdot \omega_0 \cdot S}{S^2 + \frac{\omega_0}{Q} S + \omega_0^2}} $$

```python
import numpy as np
from scipy.signal import TransferFunction
from pytc2.sistemas_lineales import bodePlot, pzmap, GroupDelay
import matplotlib.pyplot as plt
```

```python
# Parámetros normalizados
Q = 20
K = 10  # ya que R4 = 9, R5 = 1
num_norm = [0, K/Q, 0]  # s·K/Q
den_norm = [1, 1/Q, 1]  # s^2 + s/Q + 1
num = [0, K/Q, 0]
den = [1, 1/Q, 1]

H = TransferFunction(num_norm, den_norm)

bodePlot(H, fig_id=1, filter_description='H(s) normalizado')
pzmap(H, fig_id=2, filter_description='H(s) normalizado')
GroupDelay(H, fig_id=3, filter_description='H(s) normalizado')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, bode
# Partiendo de los parámetros normalizados desnormalizamos
Q = 20
K = 10  # ya que R4 = 9, R5 = 1
freqs=np.logspace(3,6,1000)
f0=10000
w0=2*np.pi*f0
num_d=[0, (K/Q)*w0, 0]
den_d=[1, w0/Q, w0*w0]

H_d=TransferFunction(num_d,den_d)
w,mag,phase,=bode(H_d,freqs)

plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.semilogx(w/(2*np.pi),mag)
plt.title("Modulo |H(s)|")
plt.ylabel("dB")
plt.grid(which='both', linestyle='--',linewidth=0.5)

plt.subplot(2,1,2)
plt.semilogx(w/(2*np.pi),phase)
plt.title("Fase H(s)")
plt.xlabel("Hz")
plt.ylabel("Fase [º]")
plt.grid(which='both', linestyle='--',linewidth=0.5)
           
plt.tight_layout()
plt.show()
```

## Calculo de R4 y R5 para una frecuencia de $f_0 = 10\,\text{kHz}$ con una ganancia de $20\,\text{dB}$ 

**Especificaciones:**

- Frecuencia central: $f_0 = 10\,\text{kHz} \Rightarrow \omega_0 = 2\pi \cdot 10^4 = 62831.85\,\text{rad/s}$
- Capacitores: $C = C_2 = 10\,\text{nF}$
- Ganancia en $f_0$: $20\,\text{dB} \Rightarrow |H(j\omega_0)| = 10$
- Factor de selectividad: $Q = 20$
  
En el dominio normalizado:  
$R = Q = 20$

en el mundo físico:  
$R = Q \cdot Z_0 = 20 \cdot 1591.55 \approx 31.83\,\text{k}\Omega$

## La función de transferencia del circuito es:

$$H(s) = \frac{ \left( s \cdot \frac{1}{RC} \right) \cdot \left( \frac{R_4}{R_5} + 1 \right)}{s^2 + \frac{1}{CR}s + \frac{R_4}{C_2 C R_1 R_3 R_5}}$$

Evaluamos esta expresión en $s = j\omega_0$ para obtener el módulo en esa frecuencia:

$H(j\omega_0) = \frac{K \cdot j\omega_0 \cdot \frac{\omega_0}{Q}}{(j\omega_0)^2 + j\omega_0 \cdot \frac{\omega_0}{Q} + \omega_0^2}$

Calculamos el denominador:

$(j\omega_0)^2 = -\omega_0^2
\Rightarrow \text{Denominador} = -\omega_0^2 + j\omega_0 \cdot \frac{\omega_0}{Q} + \omega_0^2 = j\omega_0 \cdot \frac{\omega_0}{Q}$

Entonces la transferencia evaluada en $j\omega_0$ queda:

$|H(j\omega_0)| = \left| \frac{K \cdot j\omega_0 \cdot \frac{\omega_0}{Q}}{(j\omega_0)^2 + j\omega_0 \cdot \frac{\omega_0}{Q} + \omega_0^2} \right|$

Entonces el módulo queda:

$|H(j\omega_0)| = \left| \frac{K \cdot j\omega_0 \cdot \frac{\omega_0}{Q}}{j\omega_0 \cdot \frac{\omega_0}{Q}} \right| = \frac{K \cdot \omega_0^2 / Q}{\sqrt{(-\omega_0^2 + \omega_0^2)^2 + (\omega_0^2 / Q)^2}} = \frac{K \cdot \omega_0^2 / Q}{\omega_0^2 / Q}$


$H(j\omega_0) = \frac{K \cdot j\omega_0 \cdot \frac{\omega_0}{Q}}{j\omega_0 \cdot \frac{\omega_0}{Q}} = K$

Por lo tanto:

$\boxed{|H(j\omega_0)| = |K| = 10}$


### Cálculo de $R_4$ y $R_5$:

Para tener una ganancia de $|H(j\omega_0)| = 10$, necesitamos que:

$$K = \frac{R_4}{R_5} + 1 = 10 \Rightarrow \frac{R_4}{R_5} = 9$$

Si elegimos $R_5 = 1\,\text{k}\Omega$, entonces:

$$R_4 = 9\,\text{k}\Omega$$

## $$ \boxed{R_4 = 9\,\text{k}\Omega \quad , \quad   R_5 = 1\,\text{k}\Omega} $$

## Calculo de $ R_1 \cdot R_3 $
De la función de transferencia:

$\omega_0^2 = \frac{R_4}{C^2 R_1 R_3 R_5}
\quad \Rightarrow \quad
R_1 R_3 = \frac{R_4}{R_5 \cdot C^2 \cdot \omega_0^2}$


Este producto es el que asegura que el término independiente del denominador sea igual a $\omega_0^2 $, como requiere el diseño de un filtro pasa banda con frecuencia central $f_0 $.

Reemplazamos los valores:

- $ R_4 = 9000 $
- $ R_5 = 1000 $
- $ C = 10 \cdot 10^{-9} $
- $ \omega_0 = 2\pi \cdot 10^4 \approx 62831.85 $

$ R_1 R_3 = \frac{9000}{1000 \cdot (10 \cdot 10^{-9})^2 \cdot (62831.85)^2}
= \frac{9000}{3.9478 \cdot 10^{-4}} \approx 22.797 \cdot 10^{6}$

Si $R_1=1\text{K}\Omega \Rightarrow R_3=22.97\text{K}\Omega  $

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, bode

# Definir los valores de los componentes obtenidos analiticamente
R = 31830
R1 = 1e3  
R3 = 22797
R4 = 9e3  
R5 = 1e3 
C = 10e-9
C2 = 10e-9 

freqs=np.logspace(3,6,1000)
num=[(1/(C*R))*((R4/R5)+1),0]
den=[1,1/(C*R),R4/(C*C2*R5*R3*R1)]

H_s=TransferFunction(num,den)
w,mag,phase,=bode(H_s,freqs)

plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.semilogx(w/(2*np.pi),mag)
plt.title("Modulo |H(s)|")
plt.ylabel("dB")
plt.grid(which='both', linestyle='--',linewidth=0.5)

plt.subplot(2,1,2)
plt.semilogx(w/(2*np.pi),phase)
plt.title("Fase H(s)")
plt.xlabel("Hz")
plt.ylabel("Fase [º]")
plt.grid(which='both', linestyle='--',linewidth=0.5)
           
plt.tight_layout()
plt.show()
```
